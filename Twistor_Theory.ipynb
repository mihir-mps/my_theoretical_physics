```python
import numpy as np
import matplotlib.pyplot as plt

r_list = [0.6, 1.2, 1.8]
u_angles = np.linspace(0, 2*np.pi, 12, endpoint=False)
v_ray = np.linspace(-np.pi, np.pi, 1000)

fig = plt.figure(figsize=(8, 8), facecolor="black")
ax = fig.add_subplot(111, projection="3d")
ax.set_facecolor("black")

colors = [
    "#00BFFF",  # Deep Sky Blue
    "#1E90FF",  # Dodger Blue
    "#00E5FF"   # Electric Cyan
]

for color, r in zip(colors, r_list):
    for u in u_angles:

        denom = 1 + r * np.cos(v_ray)

        X = (r * np.cos(u) + np.cos(u + v_ray)) / denom
        Y = (r * np.sin(u) + np.sin(u + v_ray)) / denom
        Z = np.sin(v_ray) / denom

        ax.plot(
            X,
            Y,
            Z,
            color=color,
            linewidth=1.6,
            alpha=0.9
        )

ax.set_title(
    "Twistor Theory Visualized",
    color="white",
    fontsize=18,
    fontweight="bold",
    pad=20
)

ax.set_box_aspect([1, 1, 1])

ax.set_axis_off()
ax.grid(False)

ax.view_init(elev=25, azim=35)

plt.tight_layout()

plt.show()
```
